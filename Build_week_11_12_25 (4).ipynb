```python
#SCARPE ALTE PER UOMO SUPERGA
import requests
from bs4 import BeautifulSoup
import pandas as pd
BASE_URL = "https://www.superga.com/it/collections/superga-scarpe-alte-uomo"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(BASE_URL, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
products = soup.find_all("div", class_="grid-product__content")

nomi = []
colori = []


for product in products:
  
    name_el = product.find("a", class_="grid-product__link")
    if not name_el:
        continue

    full_name = name_el.get_text(strip=True)

    color_el = product.find("span", class_="colorname-title")
    colore = color_el.get_text(strip=True) if color_el else ""
    colori.append(colore)

 
    base_name = full_name.replace(colore, "").strip()
    nomi.append(base_name)

  

    print(base_name, "---", colore)
```

```python
prezzi = []
for product in products:
    prezzo = product.find("div", class_="grid-product__price").text
    prezzo = prezzo.strip()
    prezzo = prezzo.replace("€", "")
    prezzo = prezzo.replace(",", ".")
    prezzo = float(prezzo)
    prezzi.append(prezzo)
prezzi
```

SCARPE BASSE DONNA

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = "https://www.superga.com/it/collections/scarpe-basse-donna"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

nomi = []
colori = []
prezzi = []

# quante pagine vuoi prendere
last_page = 6

for page in range(1, last_page + 1):
    print(f"Scarico pagina {page}...")

    url = f"{BASE_URL}?page={page}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    products = soup.find_all("div", class_="grid-product__content")

    for product in products:

        # Nome completo
        name_el = product.find("a", class_="grid-product__link")
        if not name_el:
            continue

        full_name = name_el.get_text(strip=True)

        # Colore
        color_el = product.find("span", class_="colorname-title")
        colore = color_el.get_text(strip=True) if color_el else ""
        colori.append(colore)

        # Rimuove il colore se presente, anche se attaccato
        base_name = full_name.replace(colore, "").replace("  ", " ").strip()
        nomi.append(base_name)
    
        pr = product.find("div", class_="grid-product__price")

        if pr:
            prezzo = pr.text.strip().replace("€", "").replace(",", ".")
            prezzi.append(float(prezzo))
        else:
            prezzi.append(None)

        print(base_name, "---", colore,)
print(prezzi)
```

```python
df_Donna = pd.DataFrame(
{'marca':'SUPERGA',
 'nome':nomi,
'categoria':'Scarpe Basse',
"colore": colori,
'genere':'Donna',
"prezzo": prezzi
}
)
df_Donna
```
