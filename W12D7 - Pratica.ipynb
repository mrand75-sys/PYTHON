Esercizio Sul database titanic, effettuiamo le seguenti analisi:
• Riusciamo a calcolare quanti ponti c'erano sulla nave (prima lettera dei dati nella colonna Cabin)?
• Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco (colonna PClass) 
• Fare la stessa cosa per la colonna Survived
• Qual era la distribuzione delle tariffe (Fare)? 
• Riusciamo a vedere la distribuzione delle età dei passeggeri rispetto alla classe di imbarco con un boxplot?
• Visualizziamo un boxplot rispetto alle colonne Fare e Survived; che cose ne deduciamo?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset Titanic (ad esempio da seaborn)
df = sns.load_dataset("titanic")
print(df.head())
```

```python
df = sns.load_dataset("titanic")
df.info()
```

```python
# Estrarre la prima lettera della cabina
df['deck'] = df['cabin'].dropna().str[0]  # Changed from 'pclass' to 'cabin'
print("Ponti presenti:", df['deck'].unique())
```

```python
#Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco (colonna PClass) 
sns.countplot(data=df, x="pclass", palette="Set2")
plt.title("Numero di passeggeri per classe di imbarco")
plt.show()
```

```python
sns.countplot(data=df, x="survived", palette="Set1")
plt.title("Numero di passeggeri sopravvissuti vs non sopravvissuti")
plt.show()
```

```python
sns.histplot(df['fare'], bins=30, kde=True, color="blue")
plt.title("Distribuzione delle tariffe (Fare)")
plt.show()
```

```python
sns.boxplot(data=df, x="pclass", y="age", palette="pastel")
plt.title("Distribuzione delle età per classe di imbarco")
plt.show()
```

```python
sns.boxplot(data=df, x="survived", y="fare", palette="coolwarm")
plt.title("Distribuzione delle tariffe rispetto alla sopravvivenza")
plt.show()
```

Esercizio Scarichiamo il dataset stockdata.csv da https://github.com/plotly/datasets/blob/master/stockdata.csv e: 
• Visualizziamo l'andamento delle azioni della Apple ("AAPL") 
• Visualizziamo l'andamento di tutte le azioni del dataset 
• Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM") negli anni 2012 e 2013 
• Con un istogramma, troviamo il range di valori che le azioni di Microsoft hanno assunto più spesso

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carichiamo il dataset direttamente dal link raw
url = "https://raw.githubusercontent.com/plotly/datasets/master/stockdata.csv"
df = pd.read_csv(url)

print(df.head())

```

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/stockdata.csv"
df = pd.read_csv(url)

print(df.info())
```

```python
plt.plot(df['Date'], df['AAPL'], label='Apple (AAPL)', color='red')
```

```python
plt.figure(figsize=(12,6))
for col in df.columns[1:]:   # tutte le colonne tranne 'Date'
    plt.plot(df['Date'], df[col], label=col)

plt.xticks(rotation=45)
plt.title("Andamento di tutte le azioni")
plt.xlabel("Data")
plt.ylabel("Prezzo")
plt.legend()
plt.show()
```

```python
# Convertiamo la colonna Date in formato datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filtriamo solo 2012 e 2013
mask = (df['Date'].dt.year >= 2012) & (df['Date'].dt.year <= 2013)
df_filtered = df.loc[mask]

plt.figure(figsize=(10,5))
plt.plot(df_filtered['Date'], df_filtered['MSFT'], label='Microsoft (MSFT)', color='blue')
plt.plot(df_filtered['Date'], df_filtered['IBM'], label='IBM', color='green')
plt.xticks(rotation=45)
plt.title("Andamento MSFT e IBM (2012-2013)")
plt.xlabel("Data")
plt.ylabel("Prezzo")
plt.legend()
plt.show()
```
