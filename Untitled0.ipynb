Scriviamo un programma che chiede in input all'utente una stringa e visualizza i primi 3 caratteri, seguiti da 3 punti di sospensione e quindi gli ultimi 3 caratteri, similmente all'esercizio della lezione scorsa. Stavolta facciamo attenzione a tutti i casi particolari, ovvero implementare soluzioni ad hoc per stringhe di lunghezza inferiore a 6 caratteri.

```python
# Chiediamo all'utente di inserire una stringa
testo = input("Inserisci una stringa: ")

# Controlliamo la lunghezza della stringa
lunghezza = len(testo)

if lunghezza >= 6:
    # Caso normale: almeno 6 caratteri
    risultato = testo[:3] + "..." + testo[-3:]
elif lunghezza > 3:
    # Caso intermedio: tra 4 e 5 caratteri
    risultato = testo[:3] + "..." + testo[-(lunghezza-3):]
else:
    # Caso speciale: 3 o meno caratteri
    risultato = testo

print("Risultato:", risultato)
```

Esercizio
Memorizza e stampa tutti i fattori di un numero dato in input (cioè tutti i numeri primi che, moltiplicati tra di loro, danno il numero di partenza). Esempio: • input: 150 • output: [2, 3, 5, 5]

```python
# Chiediamo all'utente di inserire un numero
numero = int(input("Inserisci un numero: "))

# Lista per memorizzare i fattori primi
fattori = []

# Divisore iniziale
divisore = 2

# Ciclo finché il numero è maggiore di 1
while numero > 1:
    if numero % divisore == 0:
        fattori.append(divisore)   # aggiungiamo il fattore
        numero = numero // divisore  # riduciamo il numero
    else:
        divisore += 1  # proviamo il prossimo divisore

print("Fattori primi:", fattori)
```

Calcolare (ma non stampare) le prime N potenze di K; ognuna di esse andrà memorizzata in coda a una lista. Alla fine, stampare la lista risultante. Proviamo con diversi valori di K, oppure facciamola inserire all'utente. Realizzare due versioni: • con un ciclo while, • con un ciclo for.

```python
# Chiediamo all'utente di inserire N e K
N = int(input("Quante potenze vuoi calcolare? "))
K = int(input("Di quale numero vuoi calcolare le potenze? "))

# Lista vuota per memorizzare le potenze
potenze = []

# Inizializziamo l'esponente
esponente = 0

# Ciclo while
while esponente < N:
    potenze.append(K ** esponente)  # calcoliamo e aggiungiamo la potenza
    esponente += 1  # incremento dell'esponente

# Stampiamo la lista finale
print("Lista delle potenze:", potenze)
```

```python
# Chiediamo all'utente di inserire N e K
N = int(input("Quante potenze vuoi calcolare? "))
K = int(input("Di quale numero vuoi calcolare le potenze? "))

# Lista vuota per memorizzare le potenze
potenze = []

# Ciclo for
for esponente in range(N):
    potenze.append(K ** esponente)

# Stampiamo la lista finale
print("Lista delle potenze:", potenze)
```

Esercizio
Abbiamo una lista con i guadagni degli ultimi 12 mesi: guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50] usando un costrutto for, calcolare la media dei guadagni e stamparla a video.

```python
# Lista dei guadagni degli ultimi 12 mesi
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

# Variabile per sommare i guadagni
somma = 0

# Ciclo for per sommare tutti i valori
for valore in guadagni:
    somma += valore

# Calcoliamo la media
media = somma / len(guadagni)

# Stampiamo il risultato
print("La media dei guadagni è:", media)
```

Abbiamo una lista di parole: parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere", "Semestre", "Esteta", "Sosta", "Orpello", "Abete", "Orologio", "Cesta", "Ermellino"] stampiamo, per ogni parola, quante volte appare la lettera "e"; facciamo attenzione al fatto che appare sia maiuscola che minuscola.

```python
# Lista di parole
parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere",
          "Semestre", "Esteta", "Sosta", "Orpello", "Abete", "Orologio",
          "Cesta", "Ermellino"]

# Ciclo for per ogni parola
for parola in parole:
    # Convertiamo la parola in minuscolo per uniformare
    conteggio = parola.lower().count("e")
    print(parola, "->", conteggio)
```

Esercizio
Abbiamo un dizionario che assegna ad ogni proprietario la sua auto: dizionario_auto = {"Ada": "Punto", "Ben": "Multipla", "Charlie": "Golf", "Debbie": "107"} Con un ciclo, e usando il metodo .values(), stampiamo a video tutte le auto che non sono una Multipla.

```python
# Dizionario con proprietari e auto
dizionario_auto = {
    "Ada": "Punto",
    "Ben": "Multipla",
    "Charlie": "Golf",
    "Debbie": "107"
}

# Ciclo for che scorre solo i valori (le auto)
for auto in dizionario_auto.values():
    if auto != "Multipla":   # condizione per escludere la Multipla
        print(auto)
```
